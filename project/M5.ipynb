```python
def check_missing_values(df, df_name):
    missing_values = df.isnull().sum()
    missing_percentage = (missing_values / len(df)) * 100
    missing_info = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
    print(f"Missing Values in {df_name}:")
    print(missing_info[missing_info['Missing Values'] > 0])
    print('\n')

# Check for missing values
check_missing_values(calendar, 'calendar.csv')
check_missing_values(sale, 'sales_train_validation.csv')
check_missing_values(sales_eva, 'sales_train_evaluation.csv')
check_missing_values(price, 'sell_prices.csv')
```

```python
def check_duplicate_rows(df, df_name):
    duplicate_rows = df.duplicated().sum()
    print(f"Duplicate Rows in {df_name}: {duplicate_rows}\n")

# 检查重复值
check_duplicate_rows(calendar, 'calendar.csv')
check_duplicate_rows(sale, 'sales_train_validation.csv')
check_duplicate_rows(sales_eva, 'sales_train_evaluation.csv')
check_duplicate_rows(price, 'sell_prices.csv')
```

```python
# 计算同比 (YoY) 变化
sales_by_category_time['YoY'] = sales_by_category_time.groupby('cat_id')['sales'].pct_change(periods=12) * 100

# 计算环比 (MoM) 变化
sales_by_category_time['MoM'] = sales_by_category_time.groupby('cat_id')['sales'].pct_change(periods=1) * 100

# 删除第一个月的数据（因为无法计算环比）
sales_by_category_time_filteredM = sales_by_category_time[sales_by_category_time['year_month'] >= '2011-01']

# 去掉第一年的数据，因为无法计算同比增长率
sales_by_category_time_filteredY = sales_by_category_time[sales_by_category_time['year_month'] > '2012-01']

# 绘制同比增长率
plt.figure(figsize=(12, 6))
sns.lineplot(x='year_month', y='YoY', hue='cat_id', data=sales_by_category_time_filteredY)
plt.title('Year-over-Year (YoY) Sales Growth by Product Category')
plt.xlabel('Year-Month')
plt.ylabel('YoY Growth Rate')
plt.axhline(y=0, color='gray', linestyle='--')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.show()

# 绘制环比增长率
plt.figure(figsize=(12, 6))
sns.lineplot(x='year_month', y='MoM', hue='cat_id', data=sales_by_category_time_filteredY)
plt.title('Month-over-Month (MoM) Sales Growth by Product Category')
plt.xlabel('Year-Month')
plt.ylabel('MoM Growth Rate')
plt.axhline(y=0, color='gray', linestyle='--')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.show()
```

**Year-on-year (YoY) sales growth**

Food (blue): In the first half of the graph, the year-on-year growth rate for food products fluctuates considerably, reaching a high of close to 60 per cent, before gradually declining and tending towards zero or even negative growth. This may reflect the fact that the food category experienced a period of high demand growth, but then gradually levelled off.

Hobbies (orange): The relatively unstable year-on-year growth trend in Hobbies, sometimes with significant negative growth (below -10%), suggests that demand for this category is likely to be more seasonal or heavily influenced by short-term market factors.

Household goods (green): Household goods maintained relatively stable growth for most of the time, although there were fluctuations, but the overall trend showed a large peak in the middle of the period, and then tended to stabilise growth.

It can be seen that individual categories have experienced large fluctuations over time, particularly in the food and housewares categories.

**Month-on-Month (MoM) sales growth**

Food (blue): The chain growth rate for food switches frequently between positive and negative fluctuations, indicating that food sales are more volatile between months. This may be related to factors such as seasonal promotions and holidays.

Hobbies (orange): The YoY changes in the Hobbies category are also more dramatic, sometimes with significant positive growth and sometimes with significant negative growth, suggesting irregular fluctuations in the demand for such products.

Household Goods (green): The chain change in Household Goods was relatively smooth, but there were also significant fluctuations at some points in time, which could mean that this category is less affected by short-term external factors, but still subject to some market volatility.

```python

# 按商品计算价格变化百分比和销量变化百分比
sales_prices_merged['price_pct_change'] = sales_prices_merged.groupby('item_id')['sell_price'].pct_change()
sales_prices_merged['sales_pct_change'] = sales_prices_merged.groupby('item_id')['sales'].pct_change()
```

```python
sales_prices_merged['price_pct_change'] = sales_prices_merged['price_pct_change'].fillna(0)
sales_prices_merged['sales_pct_change']=sales_prices_merged['sales_pct_change'].fillna(0)
```

```python
# 计算价格弹性
sales_prices_merged['price_elasticity'] = sales_prices_merged['sales_pct_change'] / sales_prices_merged['price_pct_change']

# 删除弹性无法计算的行（如价格变化为0时）
sales_prices_computed = sales_prices_merged.dropna(subset=['price_elasticity'])

# 查看弹性数据
print(sales_prices_computed[['item_id', 'wm_yr_wk', 'price_elasticity']].head())
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python

# 准备数据
X = sales_with_date.drop(['sales'], axis=1)  # 特征
y = sales_with_date['sales']  # 目标值，销售量

# 分割训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# 定义 XGBoost 模型
xg_reg = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=1000)

# 训练模型
xg_reg.fit(X_train, y_train)

# 预测未来的销售量
sales_volume_pred = xg_reg.predict(X_test)
```

```python
# 假设 price_data 包含销售价格的历史数据
X_price = sales_prices_merged.drop(['price'], axis=1)  # 特征
y_price = sales_prices_merged['price']  # 目标值，价格

# 分割训练集和测试集
X_price_train, X_price_test, y_price_train, y_price_test = train_test_split(X_price, y_price, test_size=0.2, shuffle=False)

# 使用 XGBoost 预测价格
xg_price_reg = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=1000)
xg_price_reg.fit(X_price_train, y_price_train)

# 预测未来价格
price_pred = xg_price_reg.predict(X_price_test)
```
